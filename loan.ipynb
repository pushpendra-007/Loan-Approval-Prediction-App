```python
import pandas as pd
```

```python
data = pd.read_csv('loan_approval_dataset.csv')
```

```python
data.drop(columns = ['Gender'], inplace = True)
```

```python
data.columns = data.columns.str. strip() 
```

```python
data.drop(columns = ['Dependents', 'Property_Area', 'Self_Employed', 'Loan_Status'], inplace = True)
```

```python
data.Education.unique()
```

```python
def clean_data(st) :
    st = st.strip()
    return st
```

```python
clean_data('Graduate')
```

```python
data.Education = data.Education.apply(clean_data)
```

```python
data.Education.unique()
```

```python
data['Education'] = data['Education'].replace(['Graduate', 'Not Graduate'], [1,0])
```

```python
data
```

```python
data.Married.unique()
```

```python
data.Married = data.Married.apply(clean_data)
```

```python
data.Married.unique()
```

```python
data.Married = data.Married.replace(['Yes', 'No'],(1,0))
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns = ['Credit_History'])
output_data = data['Credit_History']
```

```python
input_data
```

```python
output_data 
```

```python
x_train, x_test, y_train, y_test = train_test_split (input_data, output_data, test_size = 0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
pred_data = pd.DataFrame([['1','1','2026','2529','415','240']],columns = ['Married' ,'Education' ,'ApplicantIncome' ,'CoapplicantIncome' , 'LoanAmount' , 'Loan_Amount_Term']) 
```

```python
pred_data = scaler.transform(pred_data)
```

```python
model.predict(pred_data)
```
